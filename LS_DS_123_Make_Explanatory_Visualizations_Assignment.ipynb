```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')
df.head()
```

```python
df.dtypes
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df['timestamp'].min()
```

```python
df['timestamp'].max()
```

```python
df = df.set_index('timestamp')
```

```python
df.head()
```

```python
df['2017-08-29']
```

```python
lastday = df['2017-08-29']
lastday_filtered = lastday[lastday['category'] == 'IMDb users']
```

```python
lastday_filtered.head()
```

```python
lastday_filtered.tail()
```

```python
lastday_filtered['respondents'].plot()
```

```python
lastday_filtered['category'].value_counts()
```

```python
pct_columns = range(1,11)
```

```python
pct_columns = [f'{i}_pct' for i in range(1,11)]
```

```python
pct_columns
```

```python
final = lastday_filtered.tail(1)
final
```

```python
final[pct_columns].T
```

```python
plot_data = final[pct_columns].T
plot_data.index = range(1,11)
```

```python
plot_data
```

```python
plot_data.max() + 4 
```

```python
plt.style.use('fivethirtyeight')

ax = plot_data.plot.bar(color= 'C1', width=0.9)

plt.xlabel('Rating', fontsize=7)
plt.ylabel('Percent of total votes', fontsize=7)
plt.title('`An Inconvenient Sequel: Truth to Power` is divisive',fontsize=15 , loc='left')
plt.text(x=-1.7,y= plot_data.max() -.03, s='IMDB ratings for the fild as of Aug. 29',fontsize=10)


plt.xticks(rotation =0)
plt.yticks(range(0,10,10))

legend = ax.legend()
legend.remove()
```
